```python
def load_best_beam_captions(beam_size):
    file_name = '../../outputs/voted_captions/{}/best_beam.pickle'.format(beam_size)
    with open(file_name, 'rb') as file:
        best_beams = pickle.load(file)
    return {im_id: decode_caption(best_beams[im_id], vocab) for im_id in sorted(best_beams)}
```

```python
def load_voted_captions(beam_size, similarity):
    file_name = '../../outputs/voted_captions/{}/{}.pickle'.format(beam_size, similarity)
    with open(file_name, 'rb') as file:
        return pickle.load(file)
```

```python
beam_captions = load_best_beam_captions(10)
```

```python
str_beam_captions = { (im_id, " ".join(caption)) for (im_id, caption) in beam_captions.items()}
```

```python
beam_10_captions_file = "../../outputs/str_captions_for_analysis/beam_10.pickle" 

with open(beam_10_captions_file, 'wb') as file:
    pickle.dump(str_beam_captions, file, pickle.HIGHEST_PROTOCOL)
```
